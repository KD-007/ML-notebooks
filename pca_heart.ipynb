```python
import pandas as pd
```

```python
df = pd.read_csv("./csvs/heart.csv")
df.sample(5)
```

```python
df.shape
```

```python
df.describe()
```

```python
col = ["Age" , "RestingBP" , "Cholesterol" , "FastingBS" , "MaxHR" , "Oldpeak"]
for i in col:
    df = df[(df[i] < (df[i].mean() + 3*df[i].std())) & (df[i] > (df[i].mean() - 3*df[i].std())) ]
```

```python
df.shape
```

```python
df.describe()
```

```python
df.sample(2)
```

```python
col = ["Sex" , "ChestPainType" , "RestingECG" , "ExerciseAngina" , "ST_Slope" , ]
for i in col:
    print ( i , df[i].unique())
```

```python
df2 = df.copy()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
col = [ "ChestPainType" , "RestingECG" ,  "ST_Slope"  ]
for i in col:
    df2[i] = LabelEncoder().fit_transform(df2[i])
```

```python
df2.sample(3)
```

```python
df2 = pd.get_dummies(df2 , drop_first=True)
df2.sample(2)
```

```python
df2
```

```python
from sklearn.preprocessing import StandardScaler
st = StandardScaler()
X = st.fit_transform(df2.drop("HeartDisease" , axis=1))
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X , df2.HeartDisease)
```

```python
from sklearn.svm import SVC
sc = SVC(kernel="linear")
sc.fit(X_train , y_train)
```

```python
sc.score(X_test , y_test)
```

```python
from sklearn.decomposition import PCA
```

```python
pt = PCA(0.96)
```

```python
X_new = pt.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_new , df2.HeartDisease , test_size=0.2)
```

```python
from sklearn.svm import SVC
sc = SVC(kernel="linear")
sc.fit(X_train , y_train)
```

```python
sc.score(X_test , y_test)
```
