```python
import pandas as pd
```

```python
df = pd.read_csv("./csvs/salaries.csv")
df
```

```python
from sklearn.preprocessing import LabelEncoder
ob1 = LabelEncoder()
ob2 = LabelEncoder()
ob3 = LabelEncoder()
```

```python
df2 = df.copy()
```

```python
df2.company = ob1.fit_transform(df2.company)
df2.job = ob1.fit_transform(df2.job)
df2.degree = ob1.fit_transform(df2.degree)
df2
```

```python
df3 = pd.get_dummies(df , columns=["company" , "job" , "degree"])
df3
```

```python
df3.drop(["company_abc pharma" , "job_business manager"  ,"degree_bachelors"] , axis= 1 , inplace= True)
df3
```

```python
from sklearn.model_selection import train_test_split
X1_train , X1_test , y1_train , y1_test = train_test_split(df2[df2.columns.drop("salary_more_then_100k")] , df2["salary_more_then_100k"] , test_size=0.3)
X2_train , X2_test , y2_train , y2_test = train_test_split(df3[df3.columns.drop("salary_more_then_100k")], df3["salary_more_then_100k"] , test_size=0.3)
```

```python
from sklearn import tree
reg1 = tree.DecisionTreeClassifier()
reg2 = tree.DecisionTreeClassifier()
```

```python
reg1.fit(X1_train , y1_train)
reg2.fit(X2_train , y2_train)
```

```python
reg1.score(X1_test , y1_test)
```

```python
reg2.score(X2_test , y2_test)
```

```python
import pandas as pd
```

```python
df = pd.read_csv("./csvs/titanic.csv")
df
```

```python
df.drop(["PassengerId" ,"Name" , "SibSp" , "Parch" ,"Ticket","Embarked" ,"Cabin" ] , axis=1 , inplace=True)
df
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
df.Age = df.Age.fillna(df.Age.mean())
df
```

```python
from sklearn.preprocessing import LabelEncoder
ob1 = LabelEncoder()
df.Sex = ob1.fit_transform(df.Sex)
df
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(df[df.columns.drop("Survived")] , df.Survived , test_size=0.3)
X_train
```

```python
from sklearn import tree
reg = tree.DecisionTreeClassifier()
reg.fit(X_train , y_train)
```

```python
reg.score(X_test , y_test)
```
